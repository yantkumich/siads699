# **Import Libraries**

```python
import gradio as gr
import tensorflow as tf
import pandas as pd
import numpy as np
```

# **Load Model**

```python
model = tf.keras.saving.load_model(directory + '/699/699/model/dnn_768')
```

```python
model.predict(pd.Series(['Hello World']))
```

```python
### Load DNN Vocab
vocab_path = directory + '/699/699/model/dnn_768/dnn_vocab.txt'
vocab = []
with open(vocab_path, 'r') as fp:
    for line in fp:
        x = line[:-1]
        vocab.append(x)
```

# **Backend Function**

```python
def gen_tags(Title, Description):
  input = str(Title) + str(Description)
  prob = model.predict(pd.Series([input]))
  return [x for _, x in sorted(zip(prob[0], vocab), key=lambda pair: pair[0], reverse=True)][:10]
```

# **Launch Gradio UI**

```python
gr.Interface(fn=gen_tags, inputs=["text", "text"], outputs=["textbox"],
  title="Video Tags Generator - Not for Commerical Use",
  description="Generate Tags by Video Title and Description",
  article="¬© SIADS699 Summer 2023 Outdated YouTubers - tywy@umich.edu, yantk@umich.edu").launch(share=True)
```
